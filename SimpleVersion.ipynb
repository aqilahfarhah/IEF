```python
# dataframe opertations - pandas
import pandas as pd
# plotting data - matplotlib
from matplotlib import pyplot as plt
# time series - statsmodels 
# Seasonality decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.seasonal import seasonal_decompose 
# holt winters 
# single exponential smoothing
from statsmodels.tsa.holtwinters import SimpleExpSmoothing   
# double and triple exponential smoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
banjir = pd.read_csv('/content/Bil_banjir.csv',index_col='Tahun', parse_dates=True)
# finding shape of the dataframe
print(banjir.shape)
# having a look at the data
print(banjir.head())
# plotting the original data
banjir[['Bil Banjir']].plot(title='Floods Data')
```

```python
model = ExponentialSmoothing(endog = banjir['Bil Banjir'],trend = "add",seasonal = "add",seasonal_periods = 2).fit()
```

```python
predictions = model.forecast(steps = 31)
```

```python
banjir['Bil Banjir']['2001-01-01':].plot(figsize = (9,6))

predictions.plot()
```
