```python
#copy data to copy_data variable
copy_data = []
for i in range(len(data)):
  copy_data.append(data[i])


copy_data
```

```python
#check length of the data
data_length = len(copy_data)
data_length
```

```python
#split the strings in a 2D array copy_data based on the newline (\n) and comma (,) separators, and then re-join them using commas.
for i in range(len(copy_data)):
    copy_data[i] = [elem.strip() for elem in copy_data[i].split('\n')]
    for j in range(len(copy_data[i])):
        copy_data[i][j] = [elem.strip() for elem in copy_data[i][j].split(',')]
        copy_data[i][j] = ','.join(copy_data[i][j])

copy_data

```

```python
#copy data to full_data variable
full_data = []
for i in range(len(copy_data)):
  full_data.append(copy_data[i])


full_data
```

```python
# data_clean - remove special symbol (,][])
test = []

for i in range(len(full_data)):
    for j in range(1, len(full_data[i])):
  
        # test.append(full_data[i][j])
        pattern = r"[,|[|]|'|]"
        text = re.sub(pattern, " " , full_data[i][j])
        full_data[i][j] = text + ' '
full_data 
```

```python
#split data by spacing
for i in range(len(full_data)):
    for j in range(1, len(full_data[i])):
        words = re.split(r"\s+", full_data[i][j])
        full_data[i][j] = words  # append the list of words to the row
full_data
```

```python
#convert multidimentional to 2d array
result = []
for sublist in full_data:
    if len(sublist) >= 2 and isinstance(sublist[1], list):
        result.append([sublist[0]] + sublist[1])
    else:
        result.append(sublist)

result
```

```python
#remove null values in array
data = []

# remove empty strings
for i in range(len(result)):
  data.append(list(filter(lambda x: x != '', result[i])))

data
```

```python
#duplicate row that has more than 1 date in row. example: [10/11,22/10] to become [10/11][22/10] with its representative negeri, place and so on..

new_ta = []
for row in data:
    if len(row) >= 3:
        name, *date_range, year, state = row
        for date in date_range:
            start_date, end_date = date.split('-') if '-' in date else (date, date)
            new_row = [name, f'{start_date}-{end_date}', year, state]
            new_ta.append(new_row)

new_ta
```

```python
#copy to data variable
data = []
for i in range(len(new_ta)):
  data.append(new_ta[i])


data
```

```python
# to seperate date type like: 30/10-3/11 to become 30/10,3/11
for row in data:
    for i, col in enumerate(row):
        if col.count('/') == 2 and '-' in col:
            first_date, second_date = col.split('-')
            row[i] = first_date
            row.insert(i+1, second_date)

data
```

```python
#1st step duplicate for all date type like 26-28/10.
for i in range(len(data)):
    inserted = False
    for j in range(len(data[i])):
      if data[i][j].count('-') == 1 :#check if - exist
        # print(ta[i][j])
        if not inserted: # insert only once for each sublist
              data[i].insert(j, data[i][j])
              inserted = True 

data
```

```python
#after duplicate, it will become 26-28/10, 26-28/10.. so, heres the way to convert from duplicate to 26/10, 28/10
for i in range(len(data)):
    for j in range(len(data[i])):
        if data[i][j].count('-') == 1:
            start_date = re.sub(r'-\d+/', '/', data[i][j])
            data[i][j] = start_date
            if j+1 < len(data[i]):  # check if j+1 is in range
                end_date = re.sub(r'^\d+-', '', data[i][j+1])
                data[i][j+1] = end_date
data
```

```python
# get the number of columns for each row
num_cols = [len(row) for row in data]

# print the results
print(num_cols) 
```
